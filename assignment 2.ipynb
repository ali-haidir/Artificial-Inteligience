```python
def appending(listt , node):
    
    ls = []
    if len(listt) == 0:
        return ls
    
    if len(listt) ==1:
        node.left1 = new_node(listt[0])
        ls.append(node.left1)
    if len(listt) == 2:
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        ls.append(node.left1)
        ls.append(node.left2)
        
    if len(listt) == 3:
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        node.right1 = new_node(listt[2])
        ls.append(node.left1)
        ls.append(node.left2)
        ls.append(node.right1)
    if len(listt) == 4:
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        node.right1 = new_node(listt[2])
        node.right2 = new_node(listt[3])
        ls.append(node.left1)
        ls.append(node.left2)
        ls.append(node.right1)
        ls.append(node.right2)
        
    return ls
```

```python
def BFS(root):
    #print(type(root))
    to_visit = [root]
    visited = []
    count = 1
    count2 =0
    while to_visit:
        current = to_visit.pop(0)
        if current.val in visited:
            continue
            
        visited.append(current.val)
        count2 +=1
       # print(current.val)
        
        if current.val == "I":
            print("goal state reached")
            
            print("no of iterations to reach goal state = " , count2)
            print(visited)
            break
            return
        
        
        if current.left1:
            to_visit.append(current.left1)
            count += 1
        if current.left2:
            to_visit.append(current.left2)
            count += 1
                
        if current.right1:
            to_visit.append(current.right1)
            count += 1
        if current.right2:
            to_visit.append(current.right2)
            count += 1
```
