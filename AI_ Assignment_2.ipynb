```python
matrix = [
    {('A', 3): ['D','P']},
    {('B', 4): ['V']},
    {('C', 1):['B','H','D']},
    {('D', 3): ['A','S']},
    {('E', 1):['D', 'J']},
    {('F', 3):['I' , 'U']},
    {('G', 3): ['V', 'J']},
    {('H', 3):['W']},
    {('I', 'G'): []},
    {('J',2): ['H','T']},
    {('K',3):['N']},
    {('L', 1): ['G', 'K', 'M', 'Q']},
    {('M',2): ['C','K','O','W']},
    {('N',2): ['D','L','X']},
    {('O',3): ['L']},
    {('P',4):['T']},
    {('Q',2): ['G','S']},
    {('R',3): ['C']},
    {('S',3):['D','P']},
    {('T',3):['E','Q']},
    {('U',4): ['A', 'Y']},
    {('V',1):['Q', 'U','W']},
    {('W',4):['C']},
    {('X',3):['I','U']},
    {('Y',2):['O','W']},
    
]
```

```python
class node:
    def __init__(self,val):
        self.val = val
        self.left1 = None
        self.left2 = None
        self.right1 = None
        self.right2 = None
```

```python
def new_node(val):
    new = node(val)
    return new
```

```python
def childs(node):
    for i in matrix:
        for j in i:
            if node.val == j[0]:
                return i[j]
```

```python
def linking_childs(listt , node):
    ls = []
    
    if len(listt) == 0:
        return ls
    
    if len(listt) == 1:
        node.left1 = new_node(listt[0])
        ls.append(node.left1)
        
    if len(listt) == 2:
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        ls.append(node.left1)
        ls.append(node.left2)
        
    if len(listt) == 3:
        
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        node.right1 = new_node(listt[2])
        ls.append(node.left1)
        ls.append(node.left2)
        ls.append(node.right1)
        
    if len(listt) == 4:
        node.left1 = new_node(listt[0])
        node.left2 = new_node(listt[1])
        node.right1 = new_node(listt[2]) 
        
        node.right2 = new_node(listt[3])
        ls.append(node.left1)
        ls.append(node.left2)
        ls.append(node.right1)
        ls.append(node.right2)
        
    return ls
        
```

```python
def tree(state):
    root = new_node(state)
    temp = root
    
    queue = [root]
    seen =[]
    
    while queue:
        temp = queue.pop(0)
        print("current ==  ", temp.val)
        
        if temp.val in seen:
            continue                                      #skiping nodes that are already in seen 
            
        if temp.val not in seen:
            seen.append(temp.val)
            tl1 = childs(temp)
            print( "childs of " + temp.val + "  are = " , tl1)
            
            child_nodes = linking_childs(tl1 , temp)
            if child_nodes == []:
                continue
                
            if child_nodes != []:
                for i in child_nodes:
                    queue.append(i)
                    
                    
    return root
```

```python
root = tree("X")
```

```python
def BFS(root):
    
    to_visit = [root]
    visited = []
    count = 0
    
    while to_visit:
        
        current = to_visit.pop(0)
        if current.val in visited:
            continue
            
        visited.append(current.val)
        count +=1
        
        if current.val == "I":
            print( "Your goal state is reached!! ")
            print("No of iterations to reach Goal state : ", count)
            print("path = ", visited)
            break
            return
        
        
        if current.left1:
            to_visit.append(current.left1)
            
        if current.left2:
            to_visit.append(current.left2)
        
        if current.right1:
            to_visit.append(current.right1)
            
        if current.right2:
            to_visit.append(current.right2)
```

```python
def game(state):            # thats it 
                            #thanks alot sir!!!
    root = tree(state)
    BFS(root)
```
