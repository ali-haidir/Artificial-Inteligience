```python
Matrix = [
    "#",  "#",  "#",  "#",  "#",      "#",
    "#" ,"Z", "A" , "E" , "B" , "W",   "#",
    "#" , "H" , "F" , "G" , "X" , "F",  "#",
    "#" , "Q" , "R" , "S" , "L" , "M",  "#",
    "#" , "G" ,"H" ,"T" , "O" ,"R",    "#",
    "#" ,"D" , "I" , "J", "S" ,"C" ,   "#",
    "#",     "#",    "#",   "#"  ,"#"   ,"#"
]
```

```python
class Node:
    
    def __init__(self, position:(), parent:()):
        self.position = position
        self.parent = parent
        self.g = 0 
        self.h = 0 
        self.f = 0 
    
    def __eq__(self, other):
        return self.position == other.position
    
    def __lt__(self, other):
         return self.f < other.f
    
    def __repr__(self):
        return ('({0},{1})'.format(self.position, self.f))
```

```python
def astar_search(matrix, start, end):

    openn = []
    closed = []

    start_node = Node(start, None)
    goal_node = Node(end, None)

    openn.append(start_node)
    

    while len(openn) > 0:

        openn.sort()

        current_node = openn.pop(0)

        closed.append(current_node)
        

        if current_node == goal_node:
            path = []
            while current_node != start_node:
                path.append(current_node.position)
                current_node = current_node.parent

                
            return path[::-1]

        (x, y) = current_node.position
        
        neighbors = [(x-1, y), (x+1, y), (x, y-1), (x, y+1)]
        
        for nextt in neighbors:
        
            matrix_value = matrix.get(nextt)
        
            if(matrix_value == '#'):
                continue
        
            neighbor = Node(nextt, current_node)
        
            if(neighbor in closed):
                continue
            # Generate heuristics (Manhattan distance)
            neighbor.f = huristic()
            # Check if neighbor is in openn list and if it has a lower f value
            if(add_to_openn(openn, neighbor) == True):
          
                openn.append(neighbor)
    
    return None

def huristic():
    
    
    neighbor.g = abs(neighbor.position[0] - start_node.position[0]) + abs(neighbor.position[1] - start_node.position[1])
    neighbor.h = abs(neighbor.position[0] - goal_node.position[0]) + abs(neighbor.position[1] - goal_node.position[1])
    neighbor.f = neighbor.g + neighbor.h
    
    return neighbor.f

def add_to_openn(openn, neighbor):
    for node in openn:
        if (neighbor == node and neighbor.f >= node.f):
            return False
    return True

```
